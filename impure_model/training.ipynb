```python
input_img = Input(shape=(150,150,3))
x = Conv2D(64, (3,3), activation='relu')(input_img)
x = MaxPooling2D(pool_size=(2,2))(x)
x = Conv2D(64, (3,3), activation='relu')(x)
x = MaxPooling2D(pool_size=(2,2))(x)
x = Flatten()(x)
image_model = Model(inputs=input_img, outputs=x)

input_int = Input(shape=(150,))
y = Dense(128, activation='relu')(input_int)
intensity_model = Model(inputs=input_int, outputs=y)

merged = concatenate([image_model.output, intensity_model.output])
merged = Dense(128, activation='relu')(merged)
merged = Dense(2, activation='softmax')(merged)

model = Model(inputs=[image_model.input, intensity_model.input], outputs=merged)
```

```python
model.compile(optimizer='adam', loss= 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
y = np.array(y)
model.fit([X,scaled_intensity],y, epochs=10, validation_data=([X,scaled_intensity],y))
```

```python
model.save('64x2-CNN_impure_scaled_via.model')
```
