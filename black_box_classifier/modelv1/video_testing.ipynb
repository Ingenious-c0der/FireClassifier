```python
import cv2
from keras.models import load_model
```

```python
vid_path = "..//test_videos//candle-4101.mp4" 
CATEGORIES = ["FIRE", "FLAME"] 
def prep_frame(frame):
    frame = cv2.resize(frame, (150, 150))
    frame = frame.reshape(-1, 150,150, 3)
    return frame
```

```python
video = cv2.VideoCapture(vid_path)
model = load_model("64x2-CNN.model")
while video.isOpened():
    ret, frame = video.read()
    if ret:
        print(CATEGORIES[int(model.predict([prep_frame(frame)])[0][1])])  
    else:
        break
```
